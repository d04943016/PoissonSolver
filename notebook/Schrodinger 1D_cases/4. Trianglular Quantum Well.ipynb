# Two Triangle Wells

```python
""" setting """
m = 1
V0 = 0
V1 = -10
h_bar = 1

x = np.linspace(-20, 20, 1000) # cm
V_fun = lambda x: potential.triangle_potential(x=x, x_range= [[-9, -3], [3, 9]], V0=V0, V1 = V1)
```

```python
""" solve and plot """
E, Psis = solver1D.solve_Schrodinger_eq(x, V_fun, num_eigenvalues = 11, m=m, h_bar=h_bar)
Psis = solver1D.normalize_psi_max(Psis, factors=0.5)

_, ax = plt.subplots(1, 3, figsize=(8*3, 6))
solver1D.plot_wavefunctions_and_potential(x, E, Psis, V_fun(x), labels = [f'$E_{i}$ = {E[i]:.2f}' for i in range(E.size)], ax = ax[0])
solver1D.plot_wavefunctions_and_potential(x, [E[2]], [Psis[2]], V_fun(x), labels = [f'$E_2$',], ax = ax[1], title = f'bonding / E = {E[2]:.2f}')
solver1D.plot_wavefunctions_and_potential(x, [E[3]], [Psis[3]], V_fun(x), labels = [f'$E_3$',], ax = ax[2], title = f'anti-bonding / E = {E[3]:.2f}')
```

Simiar to the rectangular quantum well notebook cases, the energy level of two triangle quantum well is the same as in the single well but with degeneracy = 2. 

# Two Triangle Wells / Side by side

```python
""" setting """
m = 1
V0 = 0
V1 = -10
h_bar = 1

x = np.linspace(-20, 20, 1000) # cm
V_fun = lambda x: potential.triangle_potential(x=x, x_range= [[-6, 0], [0, 6]], V0=V0, V1 = V1)
```

```python
""" plot potential """
solver1D.plot_potential(x, V_fun(x) )
```

```python
""" solve and plot """
E, Psis = solver1D.solve_Schrodinger_eq(x, V_fun, num_eigenvalues = 10, m=m, h_bar=h_bar)
Psis = solver1D.normalize_psi_max(Psis, factors=0.5)

_, ax = plt.subplots(2, 3, figsize=(8*3, 6*2))
ax = ax.flatten()
solver1D.plot_wavefunctions_and_potential(x, E, Psis, V_fun(x), labels = [f'$E_{i}$ = {E[i]:.2f}' for i in range(E.size)], ax = ax[0])
solver1D.plot_wavefunctions_and_potential(x, [E[2]], [Psis[2]], V_fun(x), labels = [f'$E_2$',], ax = ax[1], title = f'bonding / E = {E[2]:.2f}')
solver1D.plot_wavefunctions_and_potential(x, [E[3]], [Psis[3]], V_fun(x), labels = [f'$E_3$',], ax = ax[2], title = f'anti-bonding / E = {E[3]:.2f}')
solver1D.plot_wavefunctions_and_potential(x, [E[8]], [Psis[8]], V_fun(x), labels = [f'$E_8$',], ax = ax[3], title = f'bonding / E = {E[8]:.2f}')
solver1D.plot_wavefunctions_and_potential(x, [E[9]], [Psis[9]], V_fun(x), labels = [f'$E_9$',], ax = ax[4], title = f'anti-bonding / E = {E[9]:.2f}')
```

We reduce the distance between two wells. In the lower energy levels, the energy levels are the same as in the single quantum well case. Interestingly, the energy level splits when energy is high! When energy level is high, it also indicate the kinetic energy is high, leading to longer penetrating length into the potential well. Besides, because the distance between two well is close, the penetrated wave function would also perceived the other well. Or you can said that the wave function in two well would interact with each other and cause the total wave function distorted. In bonding case, the wave function in well 1 and well 2 are with the same sign, leading to symmetric total wave function. In contrast, in the anti-bonding case, the sign of each wave function would be different (+/-), leading to anti-symmetric wave function. Another interest thing is that the gradient of symmetric total wave function would be smaller than that of anti-symmetric total wave function. Besides, the gradient of wave function corresponds to the momentum and the kinetic energy. This shows that the symmetric total wave function would have smaller kinetic energy than anti-symmetric total wave function and the original non-interaction energy level would between these two values. However, this phenomenon is hard to be perceived in lower energy level because the the well distance is longer in small energy level.

# Two Triangle Wells / Merged

```python
""" setting """
m = 1
V0 = 0
V1 = -10
h_bar = 1

x = np.linspace(-20, 20, 1000) # cm
V_fun = lambda x: potential.triangle_potential(x=x, x_range= [[-5, 1], [-1, 5]], V0=V0, V1 = V1)
```

```python
""" plot potential """
solver1D.plot_potential(x, V_fun(x) )
```

```python
""" solve and plot """
E, Psis = solver1D.solve_Schrodinger_eq(x, V_fun, num_eigenvalues = 10, m=m, h_bar=h_bar)
Psis = solver1D.normalize_psi_max(Psis, factors=0.5)

_, ax = plt.subplots(1, 3, figsize=(8*3, 6))
solver1D.plot_wavefunctions_and_potential(x, E, Psis, V_fun(x), labels = [f'$E_{i}$ = {E[i]:.2f}' for i in range(E.size)], ax = ax[0])
solver1D.plot_wavefunctions_and_potential(x, [E[2]], [Psis[2]], V_fun(x), labels = [f'$E_2$',], ax = ax[1], title = f'bonding / E = {E[2]:.2f}')
solver1D.plot_wavefunctions_and_potential(x, [E[3]], [Psis[3]], V_fun(x), labels = [f'$E_3$',], ax = ax[2], title = f'anti-bonding / E = {E[3]:.2f}')
```

In this case, we further merge two wells. It would be much more obvious that the first energy level is still degenerate becuase of the distance between the well. However, the other degenerated states (in previous cases) are fully split. The state would split into one bonding (symmetric) state and anit-bonding (anti-symmetric) state. When the energy become much higher, the wave function would be more like a sine or cosine functions, showing that the two well are more like a single well becuase the deeps of these two wells are too small comparing to the kinetic energy. The wave function would only perceive the most left and most right potential edge.
