# Simple Harmonic Oscillator (SHO)

```python
# 
m = 1
w0 = 0.1
h_bar = 1

x = np.linspace(-15, 15, 1000) # cm
V_fun = lambda x: potential.simple_harmonic_oscillator_potential(x, k = m * w0**2, x0 = 0.0)
```

```python
""" plot potential """
solver1D.plot_potential(x, V_fun(x) )
```

```python
E, Psis = solver1D.solve_Schrodinger_eq(x, V_fun, num_eigenvalues = 10, m=m, h_bar=h_bar)
Psis = solver1D.normalize_psi_max(Psis, factors=h_bar*w0/3)
solver1D.plot_wavefunctions_and_potential(x, E, Psis, V_fun(x), labels = [f'$E_{i}$ = {E[i]:.2f}' for i in range(E.size)])
```

The energy difference between two adjacent energy level in SHO is equal to $\hbar w_0$ ($E = (n + \frac{1}{2})\hbar w_0$)
