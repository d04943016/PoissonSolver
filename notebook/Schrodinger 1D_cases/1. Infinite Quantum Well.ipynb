# Infinite Quantum Well

```python
""" setting """
m = 1
V0 = 0
V1 = -10
h_bar = 1

x = np.linspace(-15, 15, 1000) # cm
V_fun = lambda x: np.zeros_like(x)
```

```python
""" plot potential """
solver1D.plot_potential(x, V_fun(x) )
```

```python
""" solve and plot """
E, Psis = solver1D.solve_Schrodinger_eq(x, V_fun, num_eigenvalues = 10, m=m, h_bar=h_bar)
Psis = solver1D.normalize_psi_max(Psis, factors=E[1] - E[0])

_, ax = plt.subplots(1, 2, figsize=(8*2, 6))
solver1D.plot_wavefunctions_and_potential(x, E, Psis, V_fun(x), labels = [f'$E_{i}$ = {E[i]:.2f}' for i in range(E.size)], ax = ax[0])
solver1D.plot_wavefunctions_and_potential(x, [E[0], E[1], E[2]], [Psis[0], Psis[1], Psis[2]], V_fun(x), labels = [f'$E_0$',f'$E_1$',f'$E_2$'], ax = ax[1])
```

```python
solver1D.plot_kinetic_energy_and_potential_energy_at_x(x, Psis[0:2], V_fun(x), h_bar=h_bar, m=m)
```
